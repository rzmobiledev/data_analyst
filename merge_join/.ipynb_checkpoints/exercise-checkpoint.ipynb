```python
import pandas as pd
```

```python
df1 = pd.read_csv('LOTR.csv')
df1
```

```python
df2 = pd.read_csv('LOTR 2.csv')
df2
```

```python
df1.merge(df2)
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer', on='FellowshipID')
```

```python
df1.join(df2, on='FellowshipID', how='outer', lsuffix='_left', rsuffix='_Right')
```

```python
index1 = df1.set_index('FellowshipID')
index1
```

```python
index2 = df2.set_index('FellowshipID')
index2
```

```python
index1.join(index2, how='outer', lsuffix='_Left', rsuffix='_Right')
```
